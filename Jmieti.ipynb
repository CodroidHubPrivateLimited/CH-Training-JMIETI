IMPORTING NECESSARY LIBRARIES


```python
import requests
import pandas as pd
```

Jmieti Faculty Url


```python
url = "https://www.jmieti.edu.in/faculty119.html"
```

```python
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}
```

```python
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
response.content
```

```python

tables = soup.find_all("table")
print("Total tables found:", len(tables))
faculty_table = tables[5]

rows = faculty_table.find_all("tr")
```

```python
faculty_data = []

for row in rows:
    cols = row.find_all("td")
    if len(cols) >= 6:
        faculty_data.append({
            "S_No": cols[0].text.strip(),
            "Name": cols[1].text.strip(),
            "Designation": cols[2].text.strip(),
            "Qualification": cols[3].text.strip(),
            "Doctorate": cols[4].text.strip(),
            "University_ID": cols[5].text.strip()
        })
```

```python
df = pd.DataFrame(faculty_data)
df.to_csv("jmieti_faculty_data.csv", index=False)
```
