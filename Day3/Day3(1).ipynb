```python
np.random.seed(6)
N = 5000
```

```python
#synthetic data
speed = np.random.normal(45, 10, N)
fatigue = np.random.uniform(0, 10, N)
traffic = np.random.uniform(0, 10, N)
visibility = np.random.normal(5, 1, N)   
break_event = np.random.poisson(3, N)
school_hour = np.random.randint(5,18,N)
weather = np.random.choice(
    ["clear", "rain", "fog"],
    size=N,
    p=[0.6, 0.25, 0.15]   
)
```

```python
speed += np.random.normal(0,5,N)
visibility += np.random.normal(0,50,N)
```

```python
outlier = np.random.choice(N,50)
speed[outlier] = np.random.uniform(100,140,50)
```

```python
type(outlier)
```

```python
df = pd.DataFrame({
             "Speed_km/h":speed,
             "fatigue_score":fatigue,
             "traffic_density":traffic,
             "visibility":visibility,
             "break_event":break_event,
             "school_timing":school_hour,
             "weather":weather
       })
```

```python
df
```

```python
risk_score = (
    0.3 * df["Speed_km/h"] +
    0.25 * df["fatigue_score"] +
    0.2 * df["traffic_density"] +
    0.15 * df["visibility"] +
    0.1 * df["break_event"] 

)
```

```python
prob = 1/ (1 + np.exp(-risk_score)) 

df["accident_risk"] = (prob>0.5).astype(int)
df
df.to_csv("School_bus_accident_data",index = False)
```
