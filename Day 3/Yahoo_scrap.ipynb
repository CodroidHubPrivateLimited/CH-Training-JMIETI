# Stock Data Analysis
This notebook downloads stock data using the `yfinance` library, explore the data, and visualize key metrics 

```python
import yfinance as yf
```

```python
# Define Stock and Date Range
stock = "AAPL"
start_date = "2022-01-01"
end_date = "2023-01-01"

# Download Stock Data
data = yf.download(stock, start=start_date, end=end_date)

data
```

```python
# Data Overview
data.info()
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot Closing Price Over Time
plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Closing Value')  
plt.title(f'{stock} Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

# Data Exploration
Exploring dataset to gain insights into the stocks performance

```python
# Check for Missing Values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# summary statistics for the numerical columns in the dataset
statistics = data.describe()
print("\nDescriptive Statistics:\n", statistics)
```

```python
# Shows the trading volume of the stock over time
plt.figure(figsize=(12,6))
plt.plot(data['Volume'], label='Trading Volume', color='orange')
plt.title(f'{stock} Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid()
plt.show()
```
