```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
r=requests.get(url)
```

```python
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
boxes=soup.find_all("div", class_="col-md-4 col-xl-4 col-lg-4")
#print(boxes)
	
```

```python
#print(len(boxes))
```

```python
names=[n.get_text(strip=True) for n in soup.find_all("a",class_="title")]
print(names)
```

```python
des=[d.text.strip() for d in soup.find_all("p",class_="description card-text")]
```

```python
price=[p.text.strip() for p in soup.find_all("h4",class_="price float-end card-title pull-right")]
rating_items=[r.text.strip() for r in soup.find_all("div",class_="ratings")]
```

```python
tablets_data = []

for n, d, p, r in zip(names, des, price, rating_items):
    tablets_data.append({
        "Name": n,
        "Description": d,
        "Price": p,
        "Rating": r
    })

(tablets_data)
```

```python
df = pd.DataFrame(tablets_data)
print(df)
```

```python
df.to_csv("tablets_data.csv",index=False)
```
